```python
#Loading/creating data
locations = [
    ("Maxim'o", (50.877160828051615, 4.7006992525313445)),
    ("Xior", (50.87649057167669, 4.700691639839767)),
    ("Taste", (50.875850541602986, 4.700192168675618)),
    ("His & Hairs", (50.87525620749695, 4.700106135276407)),
    ("Calvariekapel KU Leuven", (50.87448678827803, 4.699889625684066)),
    ("La Filosovia", (50.87409073433792, 4.700018371714718)),
    ("81", (50.87381822269834, 4.70010170607578)),
    ("Kiosk stadspark", (50.87526203165543, 4.701475911003796)),
    ("Vrijthof", (50.87890579999999, 4.7011444846213415))
]
wthrst = pd.read_csv("Data/Weather/01_Metadata_v2.csv")
wthrst
```

```python
m = folium.Map(location = (50.875850541602986, 4.700192168675618),
                zoom_start = 17,
                min_zoom = 10,
                #tiles="cartodb positron"
                )
#m
```

```python
#code adapted from https://stackoverflow.com/questions/42330802/can-i-add-a-sequence-of-markers-on-a-folium-map

l_noise_locations = folium.FeatureGroup("Noise locations")

for name, loc in locations:
    l_noise_locations.add_child(folium.Marker(location=loc,tooltip=name))

m.add_child(l_noise_locations)

l_weather_stations = MarkerCluster(name="Weather Stations").add_to(m)

for lat, lon in zip(wthrst.loc[:,"LAT"], wthrst.loc[:,"LON"]):
    l_weather_stations.add_child(folium.Marker(location = [lat, lon], icon = folium.Icon(color = "purple", icon ="tower-observation", prefix='fa')))

m.add_child(l_weather_stations)


folium.LayerControl().add_to(m)
m
```
